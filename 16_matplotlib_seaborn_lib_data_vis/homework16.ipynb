```python
#процент от общего населения
def percent(row):
    percent = row['suicides_no'] / row['population'] *100
    return f'{percent:.4f}%'
```

```python
dead = dead.groupby('country').sum().sort_values('population', ascending=False)
dead['percent'] = dead.apply(percent, axis = 1).sort_values()
dead.sort_values('percent', ascending = False)
```
