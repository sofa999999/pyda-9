### Задание 1
Скачайте с сайта https://grouplens.org/datasets/movielens/ датасет любого размера. Определите какому фильму было выставлено больше всего оценок 5.0.

```python
import pandas as pd
```

```python
data = pd.read_csv('movies.csv')
data.head()
```

```python
data2 = pd.read_csv('ratings.csv')
new = data.merge(data2)
new = new[ new.rating == 5 ]
top = new.groupby('title').sum()
top.sort_values('rating', ascending=False, inplace = True)
top.head(1)
```

### Задание 2
По данным файла power.csv посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 года. Не учитывайте в расчетах отрицательные значения quantity.

```python
filtered_countries = data[ (data['country']=='Estonia') |  (data['country']=='Latvia') |  (data['country']=='Litva')]
# filtered_countries.country.unique()
filtered_category = filtered_countries[ (filtered_countries['category']==4) | (filtered_countries['category']==12) | (filtered_countries['category']==21) ]
# filtered_category.category.unique()
filter_year = filtered_category[ (filtered_category['year']<=2010 ) & (filtered_category['year']>=2005 )]
# filter_year.year.unique()
filter_quantity = filter_year[ (filter_year['quantity'] >= 0)]
filter_quantity.head()
topchik = filter_quantity.groupby('country').sum()
topchik 
```
